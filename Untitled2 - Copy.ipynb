```python
class survey:
    def __init__(self, age, income, spending, consumed_id):
        self.age = age
        self.income = income
        self.spending = spending
        self.consumed_id = consumed_id


    def saving(self):
        saving = self.income - self.spending

        Savings_rate = (saving / self.income) * 100
        return saving, Savings_rate

    def display(self):
        print(f"Age: {self.age}, Income: {self.income}, spending: {self.spending}, consumed id: {self.consumed_id}")
        saving, Savings_rate = self.saving()
        print(f"Saving: {saving}, Saving Rate: {self.saving()}")

# Creating 3 survey objects
consumer1 = survey(25, 45000, 20000, 12934)
consumer2 = survey(30, 52000, 12000, 16789)
consumer3 = survey(22, 38000, 18000, 98746)

# Displaying all consumers
consumer1.display()
consumer2.display()
consumer3.display()
```

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand   # attribute
        self.speed = speed

    def accelerate(self):
        self.speed += 20  # method

# Child class
class SportsCar(Car):  # Inherits from Car
    def accelerate(self):
        self.speed += 50  # Overrides parent method

# Object creation and testing
car1 = Car("Toyota", 150)
car1.accelerate()
print(car1.brand, car1.speed)  # Toyota 110

car2 = SportsCar("Ferrari", 146)
car2.accelerate()
print(car2.brand, car2.speed)  # Ferrari 170

```
